```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
stores = pd.read_csv('store.csv')
```

```python
stores
```

```python
df = train.merge(stores, on='Store', how='left')
```

```python
df
```

```python
dfProm = df[df['Promo2']==1]
```

```python
dfProm
```

```python
dfProm['Date']=pd.to_datetime(dfProm['Date'])
dfProm['FormattedDate']=dfProm['Promo2SinceYear']*1000 + dfProm['Promo2SinceWeek']*10
dfProm['FormattedDate']=pd.to_datetime(dfProm['FormattedDate'] , format='%Y%W%w')
dfProm['FormattedDateLast']=dfProm['Promo2SinceYear']*1000 + (dfProm['Promo2SinceWeek']+1)*10
dfProm['FormattedDateLast']=pd.to_datetime(dfProm['FormattedDateLast'] , format='%Y%W%w')
dfProm['FormattedDateFirst']=dfProm['Promo2SinceYear']*1000 + (dfProm['Promo2SinceWeek']-1)*10
dfProm['FormattedDateFirst']=pd.to_datetime(dfProm['FormattedDateFirst'] , format='%Y%W%w')
```

```python
df['CompetitionOpenSinceYear'].value_counts()
```

```python
dfProm
```

```python
dfPromReducedScope = dfProm[dfProm['Date']> dfProm['FormattedDateFirst']]
dfPromReducedScope = dfPromReducedScope[dfProm['Date'] < dfProm['FormattedDateLast']]
```

```python
dfPromReducedScope
```

```python
dfPromReducedScope['afterProm']= (dfPromReducedScope[dfProm['Date'] > dfPromReducedScope['FormattedDate']]).astype(int)
```

```python
dfPromReducedScope[dfProm['Date'] < dfProm['FormattedDateLast']]
dfPromReducedScope = dfPromReducedScope[dfPromReducedScope['Sales']>0]
```

```python
dfPromReducedScope['before'] = (dfPromReducedScope['Date']< dfPromReducedScope['FormattedDate']).astype(int)
dfPromReducedScope['after'] = (dfPromReducedScope['Date']>= dfPromReducedScope['FormattedDate']).astype(int)
```
